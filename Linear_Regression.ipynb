```python
prices=bangkok_data['price']
def iqr_func(prices):
    q3, q1 = np.percentile(prices, [75 ,25])
    iqr = q3 - q1
    return iqr
```

```python
#function for detect outliers in data, base on IQR criterion
def outlier_func(prices):
    outlier = []
    q3, q1 = np.percentile(prices, [75 ,25])
    iqr = q3 - q1
    for i in prices :
        if (i > (q3 + 1.5 * iqr) or i < (q1 - 1.5 * iqr)):
            outlier.append(True)
        else:
            outlier.append(False)
    return outlier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python

# mapping room_type column

mapping = {'Entire home/apt': 1,'Private room': 2,'Hotel room': 3,'Shared room':4}
bangkok_data['room_type'] = bk_data['room_type'].map(mapping)
```

```python
X = bk_data.drop(['price'],axis=1)
y = bk_data['price']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
rmse   = np.sqrt(mean_squared_error(y_test,y_pred))
print('Training Score: ', reg.score(X_train, y_train))
print('Testing Score: ', reg.score(X_test, y_test))
print('RMSE: ',rmse)
```
